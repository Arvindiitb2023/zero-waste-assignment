```python
arr = np.array([[random.randint(0, 256) for _ in range(15)] for _ in range(15)])

def boundary_traversal(matrix):
    rows, cols = len(matrix), len(matrix[0])
    result = []

    # Traverse the first row
    for j in range(cols):
        result.append(matrix[0][j])

    # Traverse the last column
    for i in range(1, rows):
        result.append(matrix[i][cols - 1])

    # Traverse the last row if there are more than one row
    if rows > 1:
        for j in range(cols - 2, -1, -1):
            result.append(matrix[rows - 1][j])

    # Traverse the first column if there are more than one column
    if cols > 1:
        for i in range(rows - 2, 0, -1):
            result.append(matrix[i][0])

    return result

res = boundary_traversal(arr)
res
```

```python
def odd_even(arr):
    res = []
    for i in range(len(arr)):
        y = []
        for j in arr[i]:
            if(j%2 == 0):
                y.append(1)
            else:
                y.append(0)
        res.append(y)
    return res

res = odd_even(arr)
res
```

```python
# Create a horizontal gradient: 0 to 255 left to right
horizontal_gradient = np.tile(np.linspace(0, 255, 256, dtype=np.uint8), (256, 1))

# Create a vertical gradient: 0 to 255 top to bottom
vertical_gradient = np.tile(np.linspace(0, 255, 256, dtype=np.uint8).reshape(256, 1), (1, 256))

# Display both images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(horizontal_gradient, cmap='gray')
axs[0].set_title("Horizontal Gradient")
axs[0].axis('off')

axs[1].imshow(vertical_gradient, cmap='gray')
axs[1].set_title("Vertical Gradient")
axs[1].axis('off')

plt.show()
```

```python
img_color = cv2.imread("image.jpg")
img_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

# Flip horizontally
h_flip = cv2.flip(img_rgb, 1)

# Flip vertically
v_flip = cv2.flip(img_rgb, 0)

mirror = cv2.flip(img_rgb, 1)

# Display all three
titles = ["Horizontal Flip", "Vertical Flip", "Mirror (Horizontal Flip)"]
images = [h_flip, v_flip, mirror]

plt.figure(figsize=(12, 4))
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
def manual_rgb_to_grayscale(img):
    B, G, R = cv2.split(img)
    gray = 0.299 * R + 0.587 * G + 0.114 * B
    return gray.astype(np.uint8)

gray_img = manual_rgb_to_grayscale(img4)
plt.imshow(gray_img, cmap='gray')
plt.title("Manual Grayscale")
plt.axis('off')
plt.show()
```

```python
# Create a mask for white pixels
white_pixels = np.all(img4 == 255, axis=2)

# Count how many True values
count = np.sum(white_pixels)
print("Number of pure white pixels:", count)
```
